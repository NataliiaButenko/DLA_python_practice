–í —Ü—å–æ–º—É –¥–æ–º–∞—à–Ω—å–æ–º—É –∑–∞–≤–¥–∞–Ω–Ω—ñ –ø—Ä–æ–¥–æ–≤–∂—É—î–º–æ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –Ω–∞–±–æ—Ä–æ–º –¥–∞–Ω–∏—Ö `supermarket_sales.csv`.

0. –Ü–º–ø–æ—Ä—Ç—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É pandas —Ç–∞ –∑—á–∏—Ç–∞–π—Ç–µ –¥–∞–Ω—ñ —É –∑–º—ñ–Ω–Ω—É `df` —Ç–∏–ø—É `pandas.DataFrame`.

```python
import pandas as pd
```

```python
data_path = '../data_py/supermarket_sales.csv'
df = pd.read_csv(data_path)
```

1. –î–æ—Å–ª—ñ–¥–∏–º–æ, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª —Å—É–ø–µ—Ä–º–∞—Ä–∫–µ—Ç—É ('Branch') —î –Ω–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à–∏–º. –î–ª—è —Ü—å–æ–≥–æ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑–∞ –∫–æ–∂–Ω–∏–º —Ñ—ñ–ª—ñ–∞–ª–æ–º —ñ –≤–∏—è–≤—ñ—Ç—å, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª –º–∞—î –Ω–∞–π–≤–∏—â–∏–π.

```python
branch_profit = df.groupby('Branch')['gross income'].sum().sort_values(ascending=False)
print(branch_profit)
print(f'\n–ù–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à–∏–º —Ñ—ñ–ª—ñ–∞–ª–æ–º —î {branch_profit.idxmax()}.')
```

2. –í —è–∫–æ–º—É –º—ñ—Å—Ç—ñ –∑–Ω–∞–π—Ö–æ–¥–∏—Ç—å—Å—è —Ñ—ñ–ª—ñ–∞–ª –∑ –Ω–∞–π–≤–∏—â–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º? –ú–æ–∂–µ –≤ —Ç–æ–º—É –º—ñ—Å—Ü—ñ –Ω–∞–º —Ä–æ–∑–º—ñ—Ç–∏—Å—Ç–∏ —â–µ –æ–¥–∏–Ω –º–∞–≥–∞–∑–∏–Ω.  
–ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—å –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É Pandas.

```python
branch_profit_city = df.groupby('City')['gross income'].sum().sort_values(ascending=False)
print(branch_profit_city)
print(f'\n–§—ñ–ª—ñ–∞–ª –∑ –Ω–∞–π–±—ñ–ª—å—à–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤ –º—ñ—Å—Ç—ñ {branch_profit_city.idxmax()}')
```

3.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –∑–≤–æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—é, —è–∫–∞ –ø–æ–∫–∞–∂–µ, —Å–∫—ñ–ª—å–∫–∏ –ø–æ–∫—É–ø–æ–∫ (—ñ–Ω–≤–æ–π—Å—ñ–≤) –±—É–ª–æ –∑—Ä–æ–±–ª–µ–Ω–æ –≤ –∫–æ–∂–Ω—ñ–π –∑ —Ñ—ñ–ª—ñ–π (`Branch`) –∑–∞ —Ä—ñ–∑–Ω–∏–º–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è–º–∏ —Ç–æ–≤–∞—Ä—ñ–≤. –ó–∞–ø–∏—à—ñ—Ç—å —Ç–∞–±–ª–∏—Ü—é –≤ –∑–º—ñ–Ω–Ω—É `invoices_by_category` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –∑–º—ñ–Ω–Ω—É –Ω–∞ –µ–∫—Ä–∞–Ω.
–¶—è —Ç–∞–±–ª–∏—Ü—è –¥–æ–ø–æ–º–æ–∂–µ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –≤ —è–∫—ñ–π —Ñ—ñ–ª—ñ—ó –∫—É–ø—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∫–æ–∂–Ω–æ—ó –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.

```python
invoices_by_category = pd.pivot_table(df, index='Branch', columns='Product line', values='Invoice ID', aggfunc='count')
invoices_by_category 
```

3.2. –í–∏–∫—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –∑–º—ñ–Ω–Ω—É `invoices_by_category` –¥–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –ø—Ä–æ–≥—Ä–∞–º–Ω–æ (—Ç–æ–±—Ç–æ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç—Ä–µ–±–∞ –Ω–µ –ø—Ä–æ—Å—Ç–æ –∑–Ω–∞–π—Ç–∏ –æ—á–∏–º–∞, –∞ –≤–∏–≤–µ—Å—Ç–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∫–æ–¥—É), –≤ —è–∫–æ–º—É —Ñ—ñ–ª—ñ–∞–ª—ñ –º–∞–≥–∞–∑–∏–Ω—É (`Branch`) –Ω–∞–π–±—ñ–ª—å—à–µ —ñ–Ω–≤–æ–π—Å—ñ–≤ —ñ–∑ –ø–æ–∫—É–ø–∫–æ—é —Ç–æ–≤–∞—Ä—ñ–≤ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó "–ï–ª–µ–∫—Ç—Ä–æ–Ω–Ω—ñ –∞–∫—Å–µ—Å—É–∞—Ä–∏" (`Electronic accessories`)?


```python
best_branch = invoices_by_category['Electronic accessories'].idxmax()
print(f'–ù–∞–π–±—ñ–ª—å—à–µ —ñ–Ω–≤–æ–π—Å—ñ–≤ —ñ–∑ –ø–æ–∫—É–ø–∫–æ—é —Ç–æ–≤–∞—Ä—ñ–≤ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó "–ï–ª–µ–∫—Ç—Ä–æ–Ω–Ω—ñ –∞–∫—Å–µ—Å—É–∞—Ä–∏" –º–∞—î —Ñ—ñ–ª—ñ–∞–ª {best_branch}')
```

–ö—Ä–æ–∫ 1. –°—Ñ–æ—Ä–º—É–ª—é–π—Ç–µ –¢–†–ò –ø–∏—Ç–∞–Ω–Ω—è (–≥—ñ–ø–æ—Ç–µ–∑–∏) –¥–æ –Ω–∞—è–≤–Ω–∏—Ö –¥–∞–Ω–∏—Ö, —è–∫—ñ –¥–æ–ø–æ–º–æ–≥–ª–∏ –± –≤–∞–º –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ —â–æ, –¥–µ —Ç–∞ –∫–æ–ª–∏ –Ω–∞–π–±—ñ–ª—å—à–µ/–Ω–∞–π–º–µ–Ω—à–µ –∫—É–ø—É—é—Ç—å, –∞–±–∏ –¥–∞—Ç–∏ –Ω–∞ –æ—Å–Ω–æ–≤—ñ —Ü–∏—Ö –≥—ñ–ø–æ—Ç–µ–∑ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—ó –±—ñ–∑–Ω–µ—Å—É. –ó–≤—ñ—Å–Ω–æ –ø–∏—Ç–∞–Ω–Ω—è –º–∞—é—Ç—å –±—É—Ç–∏ –Ω–µ —Ç–∏–º–∏, –Ω–∞ —è–∫—ñ –º–∏ –≤–∂–µ –≤—ñ–¥–ø–æ–≤—ñ–ª–∏ –≤ –∑–∞–≤–¥–∞–Ω–Ω—è—Ö –º–æ–¥—É–ª—é.

–ö—Ä–æ–∫ 2. –ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ –Ω–∞ —Å–≤–æ—ó –ø–∏—Ç–∞–Ω–Ω—è –∑ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É pandas.

–ö—Ä–æ–∫ 3. –ù–∞–ø–∏—à—ñ—Ç—å, —è–∫ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ —Ü–µ –ø–∏—Ç–∞–Ω–Ω—è –º–æ–∂–µ –±—É—Ç–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–∞ –¥–ª—è –ø—Ä–∏–π–Ω—è—Ç—Ç—è –±—ñ–∑–Ω–µ—Å —Ä—ñ—à–µ–Ω—å.   
   
 –ü–∏—Ç–∞–Ω–Ω—è –º–æ–∂—É—Ç—å –±—É—Ç–∏ –±—É–¥—å-—è–∫–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ, –∞–ª–µ —ó—Ö –º–∞—î –±—É—Ç–∏ 3. –ö–æ–∂–Ω–µ –ø–∏—Ç–∞–Ω–Ω—è –æ—Ü—ñ–Ω—é—î—Ç—å—Å—è —è–∫ 1 –∑–∞–≤–¥–∞–Ω–Ω—è. –ë–µ–∑ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –î–ó –Ω–µ –ø—Ä–∏–π–º–∞—î—Ç—å—Å—è. –Ø–∫—â–æ —î –ø–∏—Ç–∞–Ω–Ω—è —â–æ–¥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è - –ø–∏—à—ñ—Ç—å —É —á–∞—Ç üôå


**–ü–∏—Ç–∞–Ω–Ω—è 1** –ß–∏ –≤—ñ–¥—Ä—ñ–∑–Ω—è—é—Ç—å—Å—è –≤–ø–æ–¥–æ–±–∞–Ω–Ω—è —â–æ–¥–æ —Ç–æ–≤–∞—Ä—ñ–≤ —É –ø–æ—Å—Ç—ñ–π–Ω–∏—Ö –∫–ª—ñ—î–Ω—Ç—ñ–≤ —Ç–∞ –∑–≤–∏—á–∞–π–Ω–∏—Ö –ø–æ–∫—É–ø—Ü—ñ–≤? 


```python
customer_preferences = pd.pivot_table(df, index='Product line', columns='Customer type', values='Invoice ID', aggfunc='count')
customer_preferences
```

–ù–∞–π–±—ñ–ª—å—à—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫—É–ø–æ–∫ —É—á–∞—Å–Ω–∏–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–∏ –ª–æ—è–ª—å–Ω–æ—Å—Ç—ñ –∑–¥—ñ–π—Å–Ω—é—é—Ç—å —É –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó Food and beverages (94). –¢–∞–∫–æ–∂ –≤–æ–Ω–∏ –∞–∫—Ç–∏–≤–Ω—ñ –≤ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö Sports and travel (87) —Ç–∞ Fashion accessories (86).
–£ –≤–∏–±–æ—Ä—ñ –∑–≤–∏—á–∞–π–Ω–∏—Ö –ø–æ–∫—É–ø—Ü—ñ–≤ –ª—ñ–¥–µ—Ä–∞–º–∏ —î Electronic accessories (92) —Ç–∞ Fashion accessories (92).

–î–ª—è —Å—Ç–∏–º—É–ª—è—Ü—ñ—ó –ø–µ—Ä–µ—Ö–æ–¥—É –∑–≤–∏—á–∞–π–Ω–∏—Ö –ø–æ–∫—É–ø—Ü—ñ–≤ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ª–æ—è–ª—å–Ω–æ—Å—Ç—ñ –º–æ–∂–Ω–∞ –∑–∞–ø—Ä–æ–ø–æ–Ω—É–≤–∞—Ç–∏ –º–∏—Ç—Ç—î–≤—É –∑–Ω–∏–∂–∫—É –Ω–∞ –ø–æ–∫—É–ø–∫—É Electronic accessories —Ç–∞ Fashion accessories –∑–∞ —É–º–æ–≤–∏ —Ä–µ—î—Å—Ç—Ä–∞—Ü—ñ—ó –≤ –ø—Ä–æ–≥—Ä–∞–º—ñ –ª–æ—è–ª—å–Ω–æ—Å—Ç—ñ –ø—Ä—è–º–æ –Ω–∞ –∫–∞—Å—ñ. –¢–∞–∫–æ–∂ –º–æ–∂–Ω–∞ –∑–∞–ø—É—Å—Ç–∏—Ç–∏ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω—ñ –∑–∞–∫—Ä–∏—Ç—ñ —Ä–æ–∑–ø—Ä–æ–¥–∞–∂—ñ —Ç–æ–≤–∞—Ä—ñ–≤ —Ç—ñ–ª—å–∫–∏ –¥–ª—è –≤–ª–∞—Å–Ω–∏–∫—ñ–≤ –∫–∞—Ä—Ç–æ–∫ –ª–æ—è–ª—å–Ω–æ—Å—Ç—ñ.

**–ü–∏—Ç–∞–Ω–Ω—è 2** –ß–∏ –≤–∏—Ç—Ä–∞—á–∞—é—Ç—å –ª—é–¥–∏ –≤–≤–µ—á–µ—Ä—ñ –≤ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É –±—ñ–ª—å—à–µ –≥—Ä–æ—à–µ–π –∑–∞ –æ–¥–∏–Ω –≤—ñ–∑–∏—Ç, –Ω—ñ–∂ –≤–¥–µ–Ω—å?

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour
df['Time_Slot'] = df['Hour'].apply(lambda x: 'Evening' if x >= 18 else 'Daytime')
avg_check_time = df.groupby('Time_Slot')['Total'].mean()
print(avg_check_time.round(2))
```

–í–≤–µ—á–µ—Ä—ñ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ —î –º–µ–Ω—à–∏–º –Ω—ñ–∂ –≤–¥–µ–Ω—å. –ó –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ—Ö –¥–æ—Å–ª—ñ–∂–¥–µ–Ω—å –º–∏ –∑–Ω–∞—î–º–æ —â–æ –≤–µ—á—ñ—Ä —î –Ω–∞–π–ø—ñ–∫–æ–≤—ñ—à–∏–º —á–∞—Å–æ–º –∑–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—é —á–µ–∫—ñ–≤ –∞ –æ—Ç–∂–µ –ª—é–¥–∏ –∞–±–æ –±–µ—Ä—É—Ç—å –º–µ–Ω—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∞–±–æ –±–µ—Ä—É—Ç—å –¥–µ—à–µ–≤—à—ñ –ø–æ–∑–∏—Ü—ñ—ó. –ú–æ–∂–Ω–∞ –≤–≤–µ—Å—Ç–∏ –≤–µ—á—ñ—Ä–Ω—ñ –∞–∫—Ü—ñ—ó –¥–ª—è —Å—Ç–∏–º—É–ª—é–≤–∞–Ω–Ω—è –ø—ñ–¥–≤–∏—â–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ —á–µ–∫–∞. –¢–∞–∫–æ–∂ —Ü–µ –º–æ–∂–µ –±—É—Ç–∏ –ø–æ–≤'—è–∑–∞–Ω–∏–º –∑ —á–µ—Ä–≥–∞–º–∏ –¥–æ –∫–∞—Å, –æ—Ç–∂–µ —Ç—Ä–µ–±–∞ –∑–±—ñ–ª—å—à–∏—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–µ—Ä—Å–æ–Ω–∞–ª—É –¥–ª—è –æ–±—Å–ª—É–≥–æ–≤—É–≤–∞–Ω–Ω—è –∞–±–æ –≤–≤–µ—Å—Ç–∏ –∫–∞—Å–∏ —Å–∞–º–æ–æ–±—Å–ª—É–≥–æ–≤—É–≤–∞–Ω–Ω—è.

**–ü–∏—Ç–∞–Ω–Ω—è 3** –í —è–∫–æ–º—É –º—ñ—Å—Ç—ñ –∂—ñ–Ω–∫–∏ –ø—Ä–∏–Ω–æ—Å—è—Ç—å –Ω–∞–π–±—ñ–ª—å—à–∏–π —Å—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫, –∞ –≤ —è–∫–æ–º—É ‚Äî —á–æ–ª–æ–≤—ñ–∫–∏?

```python
gender_profit_city = df.groupby(['City', 'Gender'])['gross income'].sum().unstack()
print(gender_profit_city.round(2))
```
